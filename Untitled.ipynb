```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

df=pd.read_csv("bmi_train.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.plot(kind='scatter', x="Weight", y="Height", s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.isnull().sum()
```

```python
x=df[["Weight"]]
y=df[["Height"]]
```

```python
model=linear_model.LinearRegression()
```

```python
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_test, y_test, label='Actual Values', color='green', marker='o')

# Plot predicted values
plt.plot(x_test, y_pred, label='Predicted Values', color='purple', marker='x')
```
